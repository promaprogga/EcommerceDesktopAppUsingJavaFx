```python
import random
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
```

```python
originalImage=cv2.imread("/content/butterfly.jpg") 
image = cv2.resize(originalImage, (380, 300))
cv2_imshow(image)
```

```python
kernel = np.array([[0, -1, 0],
                   [-1, 5,-1],
                   [0, -1, 0]])
image_sharp = cv2.filter2D(src=image, ddepth=-1, kernel=kernel)
cv2_imshow(cv2.resize(image_sharp, (380, 300)))
```
