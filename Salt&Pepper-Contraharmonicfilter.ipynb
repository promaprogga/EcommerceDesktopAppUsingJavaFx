```python
import random
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
```

```python
originalImage=cv2.imread("/content/sky.jfif", cv2.IMREAD_GRAYSCALE) 
image = cv2.resize(originalImage, (400, 300))
cv2_imshow(image)
```

```python
def SaltPeppernoise(image):
   row , col = image.shape
   number_of_pixels = random.randint(300, 10000)
   for i in range(number_of_pixels):
        y_coord=random.randint(0, row - 1)
        x_coord=random.randint(0, col - 1)
        image[y_coord][x_coord] = 255
   for i in range(number_of_pixels):
        y_coord=random.randint(0, row - 1)
        x_coord=random.randint(0, col - 1)
        image[y_coord][x_coord] = 0
   return image
orginalimg= SaltPeppernoise(image)
Img=cv2. resize(orginalimg, (400, 300))
cv2_imshow(Img)
```

```python
def contraharmonic_mean(Img, size, Q):
    num = np.power(Img, Q + 1)
    denom = np.power(Img, Q)
    kernel = np.full(size, 1.0)
    result = cv2.filter2D(num, -1, kernel) / cv2.filter2D(denom, -1, kernel)
    return result

cv2_imshow(contraharmonic_mean(Img, (3,3), 0.5))
```
